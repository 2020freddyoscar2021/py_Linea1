```python
import pandas as pd
from datetime import date, time, datetime
```

```python
df= pd.read_csv(r"20190614_030001-20190615_030000.txt", sep="|", names =['FECHA','HORA','ESTACION','EQUIPMENT_ID','DAILY_SEQ','UNIVERSAL_SEQ','ID_TIPO','TIPO','NUM_EXTERNO','NUM_INTERNO','NUM_TRX_TARJETA','SALDO_ANTERIOR','IMPORTE','SALDO_ACTUAL','DNI','FECHA_INSERCION','TITULO','PERFIL','CADUCIDAD'
])
```

```python
df.head()
```

```python
df_3Col=df[['HORA','ESTACION','NUM_EXTERNO']]
```

```python
df_3Col['HORAN'] = df_3Col['HORA'].str.extract('([0-9]+)')
df_3Col.head()
```

```python
df_3Col.groupby('HORAN').count()
```

```python
df_18pm = df_3Col[df_3Col['HORAN'] == '17']
```

```python
Col_pivot=pd.pivot_table(df_18pm,index=['ESTACION'], aggfunc='count')
Col_pivot.head()
```

```python
df_18menos = df_3Col[(df_3Col['HORAN'] < '17') | (df_3Col['HORAN'] > '17')]
df_18menos.head()
```

```python
df_18pm_df_18menos = pd.merge(df_18pm, df_18menos, on=['NUM_EXTERNO'], how='inner', indicator=True)
df_18pm_df_18menos
```

```python
df_18pm_df_18menos['LOGICO'] = ''
df_18pm_df_18menos.loc[df_18pm_df_18menos['ESTACION_x'] == df_18pm_df_18menos['ESTACION_y'], 'LOGICO'] = '0'
df_18pm_df_18menos.loc[df_18pm_df_18menos['ESTACION_x'] != df_18pm_df_18menos['ESTACION_y'], 'LOGICO'] = '1'
df_18pm_df_18menos.head()
```

```python
df_18menosFiltrado = df_18pm_df_18menos[(df_18pm_df_18menos['LOGICO'] == '1')]
df_18menosFiltrado.head()
```

```python
df_18menosFiltrado.info()
```

```python
with pd.ExcelWriter('py_salidaL1-2019-06-14-(17pm)T.xlsx') as writer:
   Col_pivot.to_excel(writer, index= True, sheet_name='estaciones')
   df_18menosFiltrado.to_excel(writer, index=False, sheet_name='basedatos')
```
