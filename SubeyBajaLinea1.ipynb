```python
import pandas as pd
from datetime import date, time, datetime
```

```python
df= pd.read_csv(r"20190614_030001-20190615_030000.txt", sep="|", names =['FECHA','HORA','ESTACION','EQUIPMENT_ID','DAILY_SEQ','UNIVERSAL_SEQ','ID_TIPO','TIPO','NUM_EXTERNO','NUM_INTERNO','NUM_TRX_TARJETA','SALDO_ANTERIOR','IMPORTE','SALDO_ACTUAL','DNI','FECHA_INSERCION','TITULO','PERFIL','CADUCIDAD'
])
```

```python
df.head()
```

```python
df_3Col=df[['HORA','ESTACION','NUM_EXTERNO']]
```

```python
df_3Col['HORAN'] = df_3Col['HORA'].str.extract('([0-9]+)')
df_3Col.head()
```

```python
df_7am = df_3Col[df_3Col['HORAN'] == '08']
df_7am.head()
```

```python
Col_pivot=pd.pivot_table(df_7am,index=['ESTACION'], aggfunc='count')
Col_pivot.head()
```

```python
df_8mas = df_3Col[df_3Col['HORAN'] > '08']
df_8mas.head()
```

```python
df_7am_df_8mas = pd.merge(df_7am, df_8mas, on=['NUM_EXTERNO'], how='inner', indicator=True)
df_7am_df_8mas
```

```python
df_7am_df_8mas['LOGICO'] = ''
df_7am_df_8mas.loc[df_7am_df_8mas['ESTACION_x'] == df_7am_df_8mas['ESTACION_y'], 'LOGICO'] = '0'
df_7am_df_8mas.loc[df_7am_df_8mas['ESTACION_x'] != df_7am_df_8mas['ESTACION_y'], 'LOGICO'] = '1'
df_7am_df_8mas.head()
```

```python
with pd.ExcelWriter('py_salidaL1-2019-06-14-(8am).xlsx') as writer:
    df_7am_df_8mas.to_excel(writer, index=False, sheet_name='basedatos')
    Col_pivot.to_excel(writer, index= True, sheet_name='estaciones')
```
